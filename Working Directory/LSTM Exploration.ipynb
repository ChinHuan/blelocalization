```python
import joblib
import tensorflow as tf
```

```python
ble = read_ble('../Data/ble.csv')
```

```python
rssi = pd.read_csv('../Data/sequenceRssi.csv')
```

```python
rssi = rssi.pivot_table(columns="scanner", values="rssi", index="time")
rssi.rename_axis(None, axis=1, inplace=True)
rssi.index = pd.to_datetime(rssi.index)
```

```python
rssi = rssi.groupby(pd.Grouper(freq='100ms')).mean()
rssi
```

```python
loc = pd.read_csv('../Data/sequenceLoc.csv')
loc['time'] = pd.to_datetime(loc['time'])
loc.set_index('time', inplace=True)
```

```python
loc = loc.groupby(pd.Grouper(freq='100ms')).mean()
loc.ffill(inplace=True)
loc
```

```python
def transform(loc, inplace=False):
    if inplace:
        loc['X'] = 24.187023 + loc['X']
        loc['Y'] = 12.252072 - loc['Y']
    else:
        transformed = pd.DataFrame()
        transformed['X'] = 24.187023 + loc['X']
        transformed['Y'] = 12.252072 - loc['Y']
        return transformed
```

```python
transform(loc, inplace=True)
loc
```

```python
data = rssi.join(loc, how='inner')
data.shape
```

```python
data = data[:-7]
data
```

```python
def impute(data):
    data[scanners] = min_max_scaling(data[scanners])
    data.update(data.rolling('1s').mean())
    data.ffill(inplace=True)
```

```python
def reg_impute(data):
    for s in scanners:
        dx = data['X'] - ble[s]['X']
        dy = data['Y'] - ble[s]['Y']
        d = np.sqrt(np.square(dy) + np.square(dx))
        reg = joblib.load('../Models/linear_distance_models/{}.joblib'.format(s))
        ss = pd.Series(reg.predict(d.values.reshape((-1, 1))))
        ss.index = data.index
        data[s].fillna(ss, inplace=True)
```

```python
impute(data)
reg_impute(data)
```

```python
data.describe()
```

```python
def create_window(dataset, win_size, start_index=0, end_index=None):
    data = []
    labels = []

    start_index = start_index + win_size
    if end_index is None:
        end_index = len(dataset)

    for i in range(start_index, end_index):
        indices = range(i-win_size, i)
        data.append(dataset[scanners].iloc[indices].values)
        labels.append(dataset[['X', 'Y']].iloc[i].values)
    return np.array(data), np.array(labels)
```

```python
epochs = 4000
batch_size = 32

model = tf.keras.Sequential([
    tf.keras.layers.LSTM(10, activation='relu', input_shape=(10, 17)),
    tf.keras.layers.Dense(2)
])

optimizer = tf.keras.optimizers.RMSprop(0.001)
model.compile(optimizer=optimizer, loss='mse')
```

```python
X_train, y_train = create_window(data, 10)
```

```python
dt = tf.data.Dataset.from_tensor_slices((X_train, y_train))
dt = dt.batch(batch_size)
```

```python
model.fit(dt, epochs=epochs)
```

```python
model.save("../Models/LSTM_basic.h5")
```
